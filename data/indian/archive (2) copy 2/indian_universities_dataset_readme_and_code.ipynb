```python
readme_text = """
# 🇮🇳 Indian Universities Dataset (2024)

This dataset provides a structured list of **Indian universities** sourced from [Wikipedia](https://en.wikipedia.org/wiki/List_of_universities_in_India). It includes information on university names, types, locations, and year of establishment.

## 📁 Columns

- **University**: Name of the university
- **Location**: City or state where the university is based
- **Type**: Central, State, Deemed, Private, etc.
- **Established**: Year the university was founded
- **Remarks**: Additional notes (if available)

"""
```
